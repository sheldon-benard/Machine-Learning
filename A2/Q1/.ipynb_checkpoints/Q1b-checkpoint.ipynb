```python
import numpy as np
from numpy import genfromtxt
from matplotlib import pyplot as plt
import math
```

```python
x = np.sort(np.random.uniform(0,10,50)).reshape(-1, 1)
y = x*np.sin(x) + np.random.normal(0,math.sqrt(0.5),50).reshape(-1, 1)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# RBF Scale = 0.1

kernel = 1.1*RBF(length_scale=0.1, length_scale_bounds=(1e-3, 1.0))
gp = GaussianProcessRegressor(kernel=kernel)
gp.fit(x,y)

x_range = np.arange(0,15.1,0.1).reshape(-1, 1)
y_pred, std = gp.predict(x_range, return_std=True)

var = std*std

# RBF Scale = 10

kernel_10 = 1.1*RBF(length_scale=10, length_scale_bounds=(1e-3, 3.0))
gp_10 = GaussianProcessRegressor(kernel=kernel_10)
gp_10.fit(x,y)

y_pred_10, std_10 = gp_10.predict(x_range, return_std=True)

var_10 = std_10*std_10

# Linear

from sklearn.gaussian_process.kernels import DotProduct

kernel_lin = DotProduct() # Defaults, no hyperparameters set
gp_lin = GaussianProcessRegressor(kernel=kernel_lin)
gp_lin.fit(x,y)

y_pred_lin, std_lin = gp_lin.predict(x_range, return_std=True)

var_lin = std_lin*std_lin

# Rational Quad

from sklearn.gaussian_process.kernels import RationalQuadratic

kernel_rq = RationalQuadratic(length_scale=0.1)
gp_rq = GaussianProcessRegressor(kernel=kernel_rq)
gp_rq.fit(x,y)

y_pred_rq, std_rq = gp_rq.predict(x_range, return_std=True)

var_rq = std_rq*std_rq

# Rational Quad + RBF

kernel_sum = RationalQuadratic() +  RBF() # Defaults, no hyperparameters set
gp_sum = GaussianProcessRegressor(kernel=kernel_sum)
gp_sum.fit(x,y)

y_pred_sum, std_sum = gp_sum.predict(x_range, return_std=True)

var_sum = std_sum*std_sum
```

```python
# RBF Initial Scale = 0.1

plt.clf()
plt.title("RBF V_0 = 0.1")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred, label="GP Mean")

plt.plot(x_range,var, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# RBF Initial Scale = 10

plt.clf()
plt.title("RBF V_0 = 10")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_10, label="GP Mean")

plt.plot(x_range,var_10, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# Linear Kernel

plt.clf()
plt.title("Linear Kernel")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_lin, label="GP Mean")

plt.plot(x_range,var_lin, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# Rat Quad

plt.clf()
plt.title("Rational Quadratic")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_rq, label="GP Mean")

plt.plot(x_range,var_rq, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# RBF + Rat Quad

plt.clf()
plt.title("Rational Quadratic + RBF")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_sum, label="GP Mean")

plt.plot(x_range,var_sum, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

```python
print(gp.log_marginal_likelihood())
print(gp_10.log_marginal_likelihood())
print(gp_lin.log_marginal_likelihood())
print(gp_rq.log_marginal_likelihood())
print(gp_sum.log_marginal_likelihood())
```

#### Comments

RBF Kernel: The RBF kernel has a much harder time modelling the function with noise. Variance remains above 12, and dips down drastically around clusters of training points (since the model is more confident at sample points)

Linear Kernel: Similar to a.ii, Bayesian linear regression is performed and a straight line model is produced. Since the underlying function is $xsin(x) + \epsilon$, a straight-line model isn't satisfactory.

Rational Quadratic: Has a worse fit than the rational quadratic for a.iii. The Rational Quadratic kernel produces a very jagged function, and it appears to be overfitting, especially at sample points.

RQ + RBF: produces a smooth model (much like in a.iv) and this seems like a good fit, even with the noise. Variance is low, so the model is pretty confident, much like a.iv.

```python
# RBF Initial Scale = 0.1

plt.clf()
plt.title("RBF V_0 = 0.1")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred, label="GP Mean")

plt.plot(x_range,var, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# RBF Initial Scale = 10

plt.clf()
plt.title("RBF V_0 = 10")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_10, label="GP Mean")

plt.plot(x_range,var_10, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# Linear Kernel

plt.clf()
plt.title("Linear Kernel")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_lin, label="GP Mean")

plt.plot(x_range,var_lin, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# Rat Quad

plt.clf()
plt.title("Rational Quadratic")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_rq, label="GP Mean")

plt.plot(x_range,var_rq, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()

# RBF + Rat Quad

plt.clf()
plt.title("Rational Quadratic + RBF")
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_sum, label="GP Mean")

plt.plot(x_range,var_sum, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

```python
print(gp.log_marginal_likelihood())
print(gp_10.log_marginal_likelihood())
print(gp_lin.log_marginal_likelihood())
print(gp_rq.log_marginal_likelihood())
print(gp_sum.log_marginal_likelihood())
```

The white noise kernel models the noise, which allows our underlying kernel, k, to model the actual underlying function and not respond to noise. As such, the kernel's impact is allowing our modelling kernel to better model the trends and not the noise.

RBF Kernel: RBF + WK produces a model similar to a.i, albeit with more variance for $10 \le x \le 15$

Linear Kernel: Similar to a.ii, Bayesian linear regression is performed and a straight line model is produced. Since the underlying function is $xsin(x) + \epsilon$, a straight-line model isn't satisfactory.

Rational Quadratic: Produces a much smoother model, as compared to the model above (with the WhiteKernel). Variance doesn't spike for $10 \le x \le 15$, similar to the RationalQuadratic model in part a.

RQ + RBF: produces a smooth model, similar to the Rational Quadratic + RBF model above, but with higher variance for $10 \le x \le 15$.
