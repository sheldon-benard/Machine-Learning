#### i) 

Fit the data using GP, with squared exponential covariance function with 

scale parameter $v_0 = 1.1$

length scale = 0.1

```python
import numpy as np
from numpy import genfromtxt
from matplotlib import pyplot as plt
import math
```

```python
x = np.sort(np.random.uniform(0,10,50)).reshape(-1, 1)
y = x*np.sin(x)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 1.1*RBF(length_scale=0.1, length_scale_bounds=(1e-3, 1.0))
gp = GaussianProcessRegressor(kernel=kernel)
gp.fit(x,y)

x_range = np.arange(0,15.1,0.1).reshape(-1, 1)
y_pred, std = gp.predict(x_range, return_std=True)

var = std*std
```

```python
plt.clf()
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred, label="GP Mean")

plt.plot(x_range,var, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

```python
var[130]
```

The variance is 0 (or nearly 0) for $x \in [0,10]$. This tells me that the model is very confident that the mean is the correct function. Sampling $\hat{f}$ from the Gaussian process would yield a tight (very similar) group of functions in this range $x \in [0,10]$. 

As we move further from our training values, i.e $10 \le x \le 15$, our variance increases (ex. $\sigma^2(13) \approx 7)$. Thus, our model is less confident with with predicting $10 \le x \le 15$, and it becomes less confident as $x \to 15$. I.e. if we sample several $\hat{f}$ and chart the function in the range $10 \le x \le 15$, we'll notice bigger differences between the sampled functions.

```python
kernel_10 = 1.1*RBF(length_scale=10, length_scale_bounds=(1e-3, 3.0))
gp_10 = GaussianProcessRegressor(kernel=kernel_10)
gp_10.fit(x,y)

y_pred_10, std_10 = gp_10.predict(x_range, return_std=True)

var_10 = std_10*std_10
```

```python
plt.clf()
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_10, label="GP Mean")

plt.plot(x_range,var_10, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

The lengthscale indicates the typical distance between turning points in the function. That being said, the GP fit of the known values ($x \in [0,10]$) seems to be equally good as compared to $l_0 = 0.1$; however, there's evidence of more variation in the GP for $x \in [0,10]$. 

Moreover, the extrapolation ($10 \le x \le 15$) starts to spike in variance later, around $x \approx 13.7$ for $l_0 = 10.0$, as compared to $x \approx 11.9$ for $l_0 = 0.1$. 

NOTE: the length_scale_bounds are still $[10^{-3}, 3.0]$

#### ii)

```python
from sklearn.gaussian_process.kernels import DotProduct

kernel_lin = DotProduct() # Defaults, no hyperparameters set
gp_lin = GaussianProcessRegressor(kernel=kernel_lin)
gp_lin.fit(x,y)

y_pred_lin, std_lin = gp_lin.predict(x_range, return_std=True)

var_lin = std_lin*std_lin
```

```python
plt.clf()
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_lin, label="GP Mean")

plt.plot(x_range,var_lin, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

A linear kernel is simply doing Bayesian linear regression, which is suitable if our underlying function was of the form $f(x_i) = \alpha + \beta*x_i + \epsilon$. This is not the case, and thus, the fit isn't suitable and is inferior when compared to the more complex RBF kernel. The model is a straight line, and it doesn't capture the curves. Whereas, the RBF kernel, given suitable length scale, can interpolate the samples extremely well.

#### iii)

```python
from sklearn.gaussian_process.kernels import RationalQuadratic

kernel_rq = RationalQuadratic(length_scale=0.1)
gp_rq = GaussianProcessRegressor(kernel=kernel_rq)
gp_rq.fit(x,y)

y_pred_rq, std_rq = gp_rq.predict(x_range, return_std=True)

var_rq = std_rq*std_rq
```

```python
plt.clf()
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_rq, label="GP Mean")

plt.plot(x_range,var_rq, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

The RationalQuadratic kernel fits the data very well, for $x \in [0,10]$, and, in this range, the variance is very low (0 or near 0). For $10 \le x \le 15$, the variance doesn't shoot up, as compared to the RBF kernel. Thus, for this range, the 95% confidence interval for the mean $\mu$ is much tighter than the RBF kernel.

Compared to the linear kernel, the RationalQuadratic kernel does a significantly better job at interpolating our samples, since this kernel can capture the curves (as opposed to the linear kernel's straight-line model).

#### iv)

```python
kernel_sum = RationalQuadratic() +  RBF() # Defaults, no hyperparameters set
gp_sum = GaussianProcessRegressor(kernel=kernel_sum)
gp_sum.fit(x,y)

y_pred_sum, std_sum = gp_sum.predict(x_range, return_std=True)

var_sum = std_sum*std_sum
```

```python
plt.clf()
plt.scatter(x,y, label="xsin(x)", c='r')
plt.plot(x_range,y_pred_sum, label="GP Mean")

plt.plot(x_range,var_sum, c='k',label="Variance $\sigma^2$")

plt.legend()
plt.show()
```

The (RationalQuadratic + RBF) kernel fits the data very well, for $x \in [0,10]$, and, in this range, the variance is very low (0 or near 0). For $10 \le x \le 15$, the variance doesn't shoot up, as compared to the RBF kernel. Thus, for this range, the 95% confidence interval for the mean $\mu$ is much tighter than the RBF kernel. However, the variance does increase more than just using the RationalQuadratic. This might be due to the influence of the RBF kernel on the RationalQuadratic kernel. The RBF kernel, for $x \ge 10$, has more variance and thus introduces more variance into the model. Thus, the 95% confidence interval for the mean $\mu$ is not as tight as using the RationalQuadratic kernel alone.

Compared to the linear kernel, the (RationalQuadratic + RBF) kernel can capture the curvature much better than the Bayesian linear regression model of the linear kernel.
